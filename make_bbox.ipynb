```python
import os
from PIL import Image
from IPython.display import display
import json
```

```python
dir_name = "example"
src_name = 'source.jpg'
trg_name = 'target.jpg'

source_img = os.path.join(dir_name, src_name)
target_img = os.path.join(dir_name, trg_name)

src_image = Image.open(source_img)
src_image = src_image.resize((512, 512))

trg_image = Image.open(target_img)
trg_image = trg_image.resize((512, 512))
```

```python
def crop_image(image, crop_box):
    cropped_image = image.crop(crop_box)
    return cropped_image

total_bbox = {}
```

```python
# Adjust the numbers within the crop box manually to ensure that the bounding box is as tight as possible around the object within the image.
src_crop_box = [190, 100, 440, 400]

cropped = crop_image(src_image, src_crop_box)
display(cropped)
```

```python
trg_crop_box = [50, 50, 500, 450]

cropped = crop_image(trg_image, trg_crop_box)
display(cropped)
```

```python
total_bbox['src']=src_crop_box
total_bbox['trg']=trg_crop_box
```

```python
with open(os.path.join(dir_name, 'crop_box.json'), 'w') as f:
    json.dump(total_bbox, f)
```
